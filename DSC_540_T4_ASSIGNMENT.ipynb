```python
import numpy as np 
import matplotlib.pyplot as plt 
  
data_points=np.linspace(-3, 3, 50) 
signal=np.sinc(data_points) 
  
```

```python
#numpy.sinc() plot
plt.plot(data_points, signal, marker = "o") 
plt.title("numpy.sinc()") 
plt.xlabel("X") 
plt.ylabel("Y") 
  
plt.show() 
```

```python
#generating Gaussian noise

noise = np.random.normal(0,0.5,50)
```

```python
#adding noise to the signal
with_noise=signal+noise
```

```python
#plot of the signal with gaussian function
plt.plot(data_points,with_noise ) 
plt.title("numpy.sinc() with gaussian noise") 
plt.xlabel("X")  
plt.ylabel("Y") 
  
plt.show() 
```

```python
#train the data with svm

from sklearn.svm import SVR

regressor = SVR(kernel = 'rbf')

regressor.fit(with_noise.reshape(-1,1),signal.reshape(-1,1))
 
 
```
